```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("D:\\CS\\Level 2 cs\\semester 1\\Data Science Methodology\\Assignment\\Assignment 1\\Assgnment1 _ dataset.csv")

data = pd.DataFrame(data)
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
data.fillna(method='pad')
```

```python
data.drop_duplicates(inplace=True)
```

```python
sns.boxplot(data['RunTime'])
```

```python
# z-Score Method

# Find The Limits 
upper_limt =data['RunTime'].mean()+3*data['RunTime'].std()
lower_limt =data['RunTime'].mean()-3*data['RunTime'].std()
print("Upper Limit ",upper_limt)
print("Lower Limit",lower_limt)
```

```python
# Find The Outliers
data.loc[(data['RunTime']>upper_limt)|(data['RunTime']<lower_limt)]
```

```python
new_data=data.loc[(data['RunTime']<=upper_limt)&(data['RunTime']>=lower_limt)]
print("Before Removing The Outliers:",len(data))
print("After Removing The Outliers:",len(new_data))
print("outliers:",len(data)-len(new_data))
```

```python
sns.boxplot(new_data['RunTime'])
```
